```python
#import statsmodels
```

```python
import statsmodels.formula.api as smf
```

```python
#train the data
```

```python
model = smf.ols('YearsExperience~Salary' , data=df).fit()
model
```

```python
#regression plot
```

```python
sns.regplot(df['YearsExperience'], df['Salary'])
```

```python
#model parameters
```

```python
model.params
```

```python
#print value for t value and p value
```

```python
print(f"The t value is {model.tvalues}\nP-value is\n {model.pvalues}")
```

```python
#r2 values

print(model.rsquared,model.rsquared_adj)
```
